## Clustering
no longer interested in prediction, looking to discover underlying similarites in data

```python
df_genes = pd.read_csv('data/nci60_data.csv',index_col=0)
```

```python
df_genes.head(5)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(2)
```

```python
df_genes_transformed = pd.DataFrame(pca.fit_transform(df_genes),columns=['PC1','PC2'])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(4)
```

```python
kmeans.fit(df_genes_transformed)
```

```python
kmeans.labels_
```

```python
df_genes_transformed['cluster'] = kmeans.labels_
```

```python
df_genes_transformed
```

```python
sns.lmplot('PC1','PC2',df_genes_transformed,hue='cluster',fit_reg=False,size=8, scatter_kws={"marker": "D", "s": 100})
```

(a) The sample size n is extremely large, and the number of predic- tors p is small. Better. Comparably, more flexible model is preferred. Because the plots span around in the low dimensional space, we have enough data to train the model, thus we can take the flexible model. However, if we can have more information on the underlying distribution of the model, such as linear or non-linear, we can have more confidence to choose between the inflexible and flexible models.

(b) The number of predictors p is extremely large, and the number of observations n is small. Worse. The inflexible model is preferred. We do not have enough observation in the high dimensional space to train the model, thus it is more essential for us to know the trend of the observations, thus inflexible model, such as linear model, can satisfy this need and effectively avoid overfitting the data.

(c) The relationship between the predictors and response is highly non-linear. better. an inflexible model by definition cannot model complex relationships. hence a flexible method would perform well.(Only if variance of error terms is low)

    5. What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?

Advantages
1. ability to model very complex relationships between input and response

Disadvantages
1. Less interpretablity
2. Are prone to Overfitting the data(modelling the error and not the signal)
3. High Variance

More flexible approaches are better when the sample size is high to train on, small number of features in contrast to observations(curse of dimensionality) and the relationship is complex and non-linear.

Less flexible approches are better when the sample size is small to train on, large number of features in contrast to observations, and a simple, usually a linear relationship between the response and predictors.
