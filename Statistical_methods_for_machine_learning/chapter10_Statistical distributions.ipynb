** Gaussian Distribution **  

    can be described using two parameters: mean and variance(spread of observations from the mean); SD is the normalized spread of observations from the mean

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
from numpy import arange
sample_space = arange(-5,5,0.001)
mean = 0
stdev = 1
pdf = norm.pdf(sample_space,mean,stdev)
plt.plot(sample_space,pdf)
```

```python
cdf = norm.cdf(sample_space)
```

```python
plt.plot(sample_space,cdf)
```

** Student's t-distribution **

uncertainty or error related to estimating population statistics for data drawn from
Gaussian distributions when the size of the sample must be taken into account


 Observations are
calculated as = x - mean(x)/ s/sqrt(n)

Where x is the observations from the Gaussian distribution, mean is the average observation
of x, S is the standard deviation and n is the total number of observations. The resulting
observations form the t-observation with (n âˆ’ 1) degrees of freedom.

```python

from scipy.stats import t
sample_space = arange(-5, 5, 0.001)
dof = len(sample_space) - 1
pdf = t.pdf(sample_space, dof)
plt.plot(sample_space,pdf)
```

```python
sample_space = arange(-5, 5, 0.001)
dof = len(sample_space) - 1
cdf = t.cdf(sample_space, dof)
plt.plot(sample_space,cdf)
```

** chi-squared distribution **

the Chi-Squared distribution is also used in statistical methods
on data drawn from a Gaussian distribution to quantify the uncertainty

An observation in a Chi-Squared distribution is calculated as the sum of k squared observations
drawn from a Gaussian distribution.


```python
from scipy.stats import chi2
```

```python
sample_space = arange(0,50,0.01)
len(sample_space)
```

```python
dof = 20
```

```python
pdf = chi2.pdf(sample_space,dof)
plt.plot(sample_space,pdf)
```

```python
cdf = chi2.cdf(sample_space,dof)
plt.plot(sample_space,cdf)
```
