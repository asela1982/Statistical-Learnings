** Line plot **

```python
from numpy import sin
import matplotlib.pyplot as plt
```

```python
x = [x*0.1 for x in range(100)]
y = sin(x)
plt.plot(x,y)
```

** Bar chart **

```python
from numpy.random import randint
```

```python
x = ['red','green','blue']
y = [randint(0,100),randint(0,100),randint(0,100)]
plt.bar(x,y)
```

** Histogram Plot **

```python
from numpy.random import randn
x = randn(1000)
```

```python
plt.hist(x, bins=25)
```

** Box and Whisker plot **

```python
x = [ randn(1000), 5*randn(1000), 10*randn(1000)]
```

```python
plt.boxplot(x)
```

** Scatter plot **

       for paired data samples - two measures(variables, columns) were recorded for a given observation such as weight and height
       
       useful to find association or correlation between two variables

```python
## draw a sample of random numbers from a standard gaussian
x = 20 + randn(1000)*100
```

```python
from numpy import mean
from numpy import var
from numpy import std
```

```python
mean(x)
```

```python
std(x)
```

```python
y = x + (100 + randn(1000)*50)
```

```python
plt.scatter(x,y)
```
