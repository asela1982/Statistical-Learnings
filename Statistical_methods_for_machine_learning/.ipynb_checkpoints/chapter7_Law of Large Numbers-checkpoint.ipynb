```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# the population
xaxis = np.arange(30,70,1)
yaxis = norm.pdf(xaxis,50,5)
plt.plot(xaxis,yaxis)
```

```python
# create sample sizes
sizes = list()
for x in range(10,20000,200):
    sizes.append(x)
```

```python
# draw random samples from the population from each of the sample sizes
from numpy.random import randn
from numpy import mean
from numpy import array
means = [mean(5*randn(size)+50) for size in sizes]
plt.scatter(sizes,array(means)-50)
```

```python
# convert a list to an array
array(means)
```
